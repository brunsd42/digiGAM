```python

def is_country(entry):
    if entry['region']['value'] == 'Aggregates' or entry['incomeLevel']['value'] == 'Aggregates':
        return False
    if not entry['capitalCity']:
        return False
    return True
    
pop_data = []
# Fetch the list of countries
countries = wbdata.get_countries()
for country in tqdm(countries):

    if is_country(country):
        iso2_code = country['iso2Code']
        #print(f"{country['name']} ({country['id']})")
        
        # Define the indicator for population data
        indicator = 'SP.POP.TOTL'
        try:
            #df = wbdata.get_dataframe(indicators, country=iso2_code)
            df = wb.get_series(indicator, country=iso2_code).reset_index()
            df['ISO2'] = iso2_code
            pop_data.append(df)
        except Exception as e:
            print(f"Error fetching data for {country['name']} ({country['id']}): {e}")    
        time.sleep(2)
    
    else: 
        print(f"skipping: {country['name']}")
```

```python
# video -> image 

import cv2
import os

def extract_frames(video_path, output_dir):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Save the frame as an image
        output_path = os.path.join(output_dir, f"frame_{frame_count}.png")
        cv2.imwrite(output_path, frame)
        frame_count += 1

    cap.release()

# Example usage
video_path = 'all_countries.mov'
output_dir = '../data/raw/we are social/'
extract_frames(video_path, output_dir)
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip
import os
from PIL import Image
```

```python
def extract_frames(video_path, output_dir):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Load the video
    clip = VideoFileClip(video_path)
    print('opened video')
    frame_count = 0

    for frame in clip.iter_frames():
        output_path = os.path.join(output_dir, f"frame_{frame_count}.png")
        frame_image = Image.fromarray(frame)
        frame_image.save(output_path)
        frame_count += 1

# Example usage
video_path = 'all_countries.mov'
output_dir = '../data/raw/we are social/'
extract_frames(video_path, output_dir)
```

```python
# Function to extract values from an image
def extract_values(image_path, image_file):
    # Open the image
    img = Image.open(image_path+image_file)
    
    # Use pytesseract to extract text
    text = pytesseract.image_to_string(img)
    
    # Extract the required values from the text
    lines = text.split('\n')
    values = {}
    for line in lines:
        if "ABKHAZIA" in line:
            values["country"] = "Abkhazia"
        elif "INDIVIDUALS USING THE INTERNET" in line:
            values["Individuals using the internet"] = int(line.split()[0].replace(',', ''))
        elif "TOTAL VS POPULATION" in line:
            values["Total vs Population"] = float(line.split()[0].replace('%', ''))
    
    return values

# List of image paths
path = "../data/raw/we are social/"
images = ["1.png", "2.png"]

# Iterate through the list of images and extract values
extracted_data = []
for image_file in images:
    extracted_data.append(extract_values(path, image_file))

print(extracted_data)
```
