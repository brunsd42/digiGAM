```python
import pandas as pd
from tqdm import tqdm
from pathlib import Path
import sys

# -------------------------------------------------------------------
# Load Lookups
# -------------------------------------------------------------------

# Add project root to import path
sys.path.append(str(Path().resolve().parents[0]))

# Path to GAM lookup
LOOKUP_XLSX = Path().resolve().parent / "helper" / "GAM_Lookup.xlsx"

# Output folder for CSV tables (docs/_static/tables)
TABLE_DIR = Path().resolve().parent / "docs" / "source" / "_static" / "tables"
TABLE_DIR.mkdir(parents=True, exist_ok=True)

# Read the Social Media lookup sheet
lookup = pd.read_excel(LOOKUP_XLSX, sheet_name="Social Media Accounts new")

# Ensure expected columns exist
required_cols = {"PlatformID", "Status", "Channel ID", "Channel Name", "ServiceID"}
missing = required_cols - set(lookup.columns)
if missing:
    raise ValueError(f"Missing expected columns in lookup: {missing}")

# -------------------------------------------------------------------
# Generate CSVs per platform × status
# -------------------------------------------------------------------

for platform in tqdm(lookup["PlatformID"].dropna().unique()):
    platform_subset = lookup[lookup["PlatformID"] == platform]

    for status in platform_subset["Status"].dropna().unique():
        status_subset = platform_subset[platform_subset["Status"] == status]
        status_subset["Channel ID"] = status_subset["Channel ID"].astype(str)
        
        # Only keep the three relevant columns
        cols = ["Channel ID", "Channel Name", "ServiceID"]
        output_df = status_subset[cols].sort_values(['ServiceID','Channel ID']).copy()

        # Target filename, e.g. "facebook_active_accounts.csv"
        filename = f"{platform}_{status}_accounts.csv"
        filepath = TABLE_DIR / filename

        # Save CSV
        output_df.to_csv(filepath, index=False)

        print(f"✓ Saved: {filepath}")
```
